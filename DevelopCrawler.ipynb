```python
from bs4 import BeautifulSoup
import requests
import os
import urllib.parse
```

```python
all_url = "https://sb-heilbronn.lmscloud.net/cgi-bin/koha/opac-search.pl?&limit=copydate%2Cst-numeric%3D-2020&sort_by=relevance&offset=0"
```

```python
resp = requests.get(all_url)
```

```python
soup = BeautifulSoup(resp.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
help(soup.find_all)
```

```python
# find all pages
n_pages = int(soup.select("#top-pages > div > ul > li:nth-child(4) > a")[0].text)
print(n_pages)
max_offset = (n_pages-1)*20
print(max_offset)
```

```python
type(soup.find_all("tr"))
```

```python
tr_list = soup.select("#bookbag_form > table > tr")
```

```python
tr_entry = tr_list[0]
```

```python
tr_entry
```

```python
#bookbag_form > table > tbody > tr:nth-child(1) > td:nth-child(1) > div > a > img
```

```python
image_url = tr_entry.select("td:nth-child(1) > div > a > img")[0].get("src")
print(image_url)
```

```python
tr_author = tr_entry.find_all("span", class_="author")[0].text
print(tr_author)
```

```python
book_uri = tr_entry.find_all("a", class_="title")[0].get("href")
book_url = urllib.parse.urljoin("https://sb-heilbronn.lmscloud.net/", book_uri)
print(book_url)
```

```python
tr_summary_book = tr_entry.find_all("span", class_="results_summary summary")[0].text
print(tr_summary_book)
```
